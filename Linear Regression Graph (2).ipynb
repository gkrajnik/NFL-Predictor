```python
#Box number 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Box number 2
df_off = pd.read_csv("yearly_team_stats_offense.csv")
df_def = pd.read_csv("yearly_team_stats_defense.csv")

df_off_filtered = df_off[(df_off['season_type'] == 'REG') & (df_off['season'] == 2024)]
df_def_filtered = df_def[(df_def['season_type'] == 'REG') & (df_def['season'] == 2024)]
```

```python
#Box number 3
X = df_off_filtered['total'].values.reshape(-1, 1)
y = df_off_filtered['win'].values
```

```python
#Box number 4
model = LinearRegression()
model.fit(X, y)
```

```python
#Box number 5
y_pred = model.predict(X)
```

```python
#Box number 6
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, y_pred, color='red', label='Linear Regression')
plt.xlabel('Offensive Yards')
plt.ylabel('Wins')
plt.title('Wins based on offensive yards for 2024 REG season')
plt.legend()
plt.savefig("wins_based_on_yards_2024.png")  # saved in your current working directory
plt.show()
```

```python
#Box number 7
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
